2. Cp --> Cantidad de Dinero

## Formulas
$Cambio = CD - CP$

```python
# Calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD >= CP:
    Cambio = CD - CP
    print("No alcanza")
else:
    cambio = CP - CD
print("Fin del programa")
print("Programa realizado pro Cepeda Darrell NL = 10 ")
```

```python
# Calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio
# si es verdadero mostrar No alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD >= CP:
    Cambio = CD - CP
    print("No alcanza")
else:
    cambio = CP - CD
    print("El cambio es: ",cambio)
print("Fin del programa")
print("Programa realizado pro Cepeda Darrell NL = 10 ")
```
